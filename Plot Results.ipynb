```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
directories = ['Static_10', 'Static_20', 'Static_30', 'uncertainty-driven', 'linear_decay', 'levy_flight', 'exponential_decay', 'adaptive_100']
```

```python
plt.figure()
for dir in directories:
    load_dir = 'data/results/' + dir
    scores = np.load(os.path.join(load_dir,'training_log_scores.npy'))
    plt.plot(scores, label=dir)
plt.legend()
plt.title('Scores')
    
```

```python
plt.figure()
for dir in directories:
    load_dir = 'data/results/' + dir
    valid = np.load(os.path.join(load_dir,'training_log_valid.npy'))
    plt.plot(valid, label=dir)
plt.legend()
plt.title('Smiles validity')
```

```python
plt.figure()
for dir in directories:
    load_dir = 'data/results/' + dir
    sa = np.load(os.path.join(load_dir,'training_log_sa.npy'))
    plt.plot(sa, label=dir)
plt.legend()
plt.title('Synthetic Accessibility score (cite paper here)')
```

```python
plt.figure()
for dir in directories:
    load_dir = 'data/results/' + dir
    novel = np.load(os.path.join(load_dir,'training_log_novel.npy'))
    plt.plot(novel, label=dir)
plt.legend()
plt.title('Novelty')
```

```python
plt.figure()
for dir in directories[3:]:
    load_dir = 'data/results/' + dir
    sigma = np.load(os.path.join(load_dir,'training_log_sigma.npy'))
    plt.plot(sigma, label=dir)
plt.legend()
```
