```python
import matplotlib.pyplot as plt
import torch
```

```python
# smaller dataset
for i, epoch_losses in enumerate(torch.load('data/Prior_losses_test')):
    losses = [l.detach().numpy() for l in epoch_losses]
    plt.plot(losses, label='Epoch: ' + str(i))
plt.legend()
plt.ylabel('Loss')
```
